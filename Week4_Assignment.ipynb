# Business Problem

A Lebanese investor wants to open a Lebanese Restaurant in Toronto. Having recently immigrated to Canada, he wants to ensure that his investment does not go to waste. The aim of this project is to analyze competing Lebanese restaurants in the different neighborhoods and find the best place to open a new restaurant.

# Data Description

To begin with, the postal code information of the the neighorhoods is required. We can grab them from here: https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M

We will rely on the scarborough dataset that we got on Week 3. The dataset consists of latitude, longitutde and the postcodes.

We will need data about the different venues in the different neighborhoods. To do so, we will rely on the Foursquare API. The API will provide us with the information we need about the venues and events in the area of interest such as: venue names, locations and menus. Using the API, we gather information about the venues in each neighborhood within a chosen radius of 100 meters. The data we retrieve will contain information on the venues within a specified distance of the longitude and latitude of the postcodes.
The information we will obtain are as follows:
Neighborhood
Neighborhood Latitude and Longitude
Venue: Name, Latitude, Longitude, Category.
